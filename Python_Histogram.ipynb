```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 10, 0.1)
y = 2*x + 1

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Straight Line Graph')
plt.show()
```

```python
x = np.random.rand(10)
y = np.random.rand(10)

plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
x = np.arange(0, 10, 0.1)
y = 2*x + 1

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

```python
x = ['A', 'B', 'C', 'D', 'E']
y = [3, 7, 5, 9, 2]

plt.bar(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar Plot')
plt.show()
```

```python
data = np.random.normal(0, 1, 1000)

plt.hist(data, bins=100)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Histogram')
plt.show()
```

```python
data = np.random.normal(0, 1, 1000)

plt.boxplot(data)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Box Plot')
plt.show()
```

```python
import seaborn as sns

data = pd.DataFrame({'A': np.random.normal(0, 1, 1000), 'B': np.random.normal(0, 1, 1000)})

sns.pairplot(data)
plt.show()
```
