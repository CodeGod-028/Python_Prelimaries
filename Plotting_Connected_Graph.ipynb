```python
# Re-import required libraries
import networkx as nx
import matplotlib.pyplot as plt

# Create subplots for connected and disconnected graphs
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Define a connected graph
connected_graph = nx.Graph()
connected_graph.add_edges_from([('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'A')])

# Define a disconnected graph
disconnected_graph = nx.Graph()
disconnected_graph.add_edges_from([('A', 'B'), ('C', 'D')])  # Two separate components

# Positions for better layout
connected_pos = nx.spring_layout(connected_graph)
disconnected_pos = nx.spring_layout(disconnected_graph)

# Draw connected graph
axes[0].set_title("Connected Graph", fontsize=14)
nx.draw(
    connected_graph,
    connected_pos,
    ax=axes[0],
    with_labels=True,
    node_color='lightgreen',
    node_size=2000,
    font_size=12,
    font_weight='bold',
    edge_color='blue',
    width=2,
)

# Draw disconnected graph
axes[1].set_title("Disconnected Graph", fontsize=14)
nx.draw(
    disconnected_graph,
    disconnected_pos,
    ax=axes[1],
    with_labels=True,
    node_color='lightcoral',
    node_size=2000,
    font_size=12,
    font_weight='bold',
    edge_color='gray',
    width=2,
)

# Display the graphs
plt.tight_layout()
plt.show()
```
