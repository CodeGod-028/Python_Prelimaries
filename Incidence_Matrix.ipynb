```python
# Importing libraries
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Create a directed graph
G = nx.DiGraph()

# Add edges (directed)
edges = [('A', 'B'), ('B', 'C'), ('C', 'A'), ('B', 'D')]
G.add_edges_from(edges)

# Draw the directed graph
plt.figure(figsize=(6, 6))
pos = nx.spring_layout(G)
nx.draw(
    G, pos, with_labels=True, node_color="lightblue", node_size=2000, font_size=12, font_weight="bold", edge_color="gray", width=2, arrowsize=20
)
nx.draw_networkx_edge_labels(
    G,
    pos,
    edge_labels={edge: f"e{i+1}" for i, edge in enumerate(edges)},
    font_color="red",
    font_size=10,
)
plt.title("Directed Graph", fontsize=14)
plt.show()

# Generate the incidence matrix
nodes = list(G.nodes)
incidence_matrix = np.zeros((len(nodes), len(edges)), dtype=int)

# Fill the incidence matrix
for edge_index, (u, v) in enumerate(edges):
    incidence_matrix[nodes.index(u), edge_index] = -1  # Outgoing edge (-1)
    incidence_matrix[nodes.index(v), edge_index] = 1   # Incoming edge (+1)

# Convert to a DataFrame for better readability
incidence_matrix_df = pd.DataFrame(
    incidence_matrix,
    index=nodes,
    columns=[f"e{i+1}" for i in range(len(edges))],
)

# Display the matrix
incidence_matrix_df
```
